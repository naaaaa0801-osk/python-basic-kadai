```python
from os import fchmod
class Human:
  def __init__(self,name,age):
    self.name=name
    self.age=age

  def check_adult(self):
    if self.age >= 20:
      print("大人です")
    else:
      print("大人ではありません")

Taro = Human ("太郎",21)
Hanako = Human ("花子",20)
Jiro = Human ("次郎",19)
Saburo = Human ("三郎",15)

family = [Taro,Hanako,Jiro,Saburo]

for info in family:
  info.check_adult();
```
