```python
class Human:
    def __init__(self):
        self.name = ""
        self.age = ""

    def set_name(self, name):
        self.name = name

    def set_age(self, age):
        self.age = age

    def show_name_and_age(self):
        print(self.name,self.age)

Taro = Human()

Taro.set_name("太郎")
Taro.set_age(5)
Taro.show_name_and_age()
```
